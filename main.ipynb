```python
#from gestion_examens import *
import import_ipynb
from gestion_examens import *
import os , json
from utils import (
    sauvegarder_donnees,
    charger_donnees,
    examens_par_date,
    etudiants_par_filiere
)
```

```python
centre = charger_donnees()

while True:
    print("\n===== MENU =====")
    print("1. Ajouter un étudiant")
    print("2. Ajouter un examen")
    print("3. Afficher les examens d'une date")
    print("4. Afficher les statistiques")
    print("5. Supprimer un examen (par matière)")
    print("6. Quitter")

    choix = input("Entrez votre choix : ")

    match choix:
        case "1":
            nom = input("Nom : ")
            numero = input("Numéro : ")
            filiere = input("Filière : ")

            etudiant = Etudiant(nom, numero, filiere)
            centre.ajouter_etudiant(etudiant)
            print("Étudiant ajouté avec succès.")

        case "2":
            matiere = input("Matière : ")
            date = input("Date : ")
            surveillant = input("Surveillant : ")

            examen = Examen(matiere, date, surveillant)
            centre.ajouter_examen(examen)
            print("Examen ajouté avec succès.")

        case "3":
            date = input("Donner la date : ")
            trouve = False

            for ex in examens_par_date(centre.examens, date):
                print(f"{ex.matiere} - {ex.date} - {ex.surveillant}")
                trouve = True

            if not trouve:
                print("Aucun examen trouvé pour cette date.")

        case "4":
            stats = centre.statistiques()
            print("Statistiques du centre :")
            print("Nombre total d'examens :", stats["total_examens"])
            print("Nombre d'étudiants inscrits :", stats["total_etudiants"])
            print("Nombre de matières uniques :", stats["matieres_uniques"])

        case "5":
            matiere = input("Donner la matière à supprimer : ")
            supprime = centre.supprimer_examen(matiere)

            if supprime:
                print("Examen supprimé avec succès.")
            else:
                print("Aucun examen trouvé avec cette matière.")

        case "6":
            sauvegarder_donnees(centre)
            print("Au revoir.")
            break

        case _:
            print("Choix invalide.")
```
